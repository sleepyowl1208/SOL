```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 100, 100)
```

```python
mark_1 = 1 * t
```

```python
mark_2 = 4 * (t - 30)
```

```python
fig, ax = plt.subplots()
plt.title("Energy generation")
plt.xlabel("time (in mins)")
plt.ylabel("distance (in km)")

ax.set_xlim([0, 60])
ax.set_ylim([0, 60])

ax.plot(t, mark_1, c="blue")
ax.plot(t, mark_2, c="red")

plt.axvline(x=40, color="purple", linestyle="--")
_ = plt.axhline(y=40, color="purple", linestyle="--")
```
