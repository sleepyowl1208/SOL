```python
import numpy as np
```

```python
x = np.array([12, 8, 29, 10])
x
```

### L1 Norm

```python
int(np.abs(12) + np.abs(8) + np.abs(29) + np.abs(10))
```

### Squared L2 Norm - Like L2 Norm but don't have the Sqrt. 

```python
(12**2) + (8**2) + (29**2) + (10**2)
```

```python
int(np.dot(x, x))
```

### Max Norm

```python
int(np.max([np.abs(12), np.abs(8), np.abs(29), np.abs(10)]))
```

### Orthogonal vectors

```python
i = np.array([1, 0])
i
```

```python
j = np.array([0, 1])
j
```

```python
int(np.dot(i, j))
```
