###L2 Norm

```python
import numpy as np
```

```python
x = np.array([12, 8, 29, 10])
x
```

```python
((12**2 + 8**2 + 29**2 + 10**2))**(1/2)
```

> This provides us with the euclidean distance or l2 norm from the origin (0,0)
> For now, the Euclidean distance form the origin is 93.12357381458253

```python
float(np.linalg.norm(x))
```

> Instead of writing ((12**2 + 8**2 + 92**2))**(1/2) explicityly, numpy comes with inbuilt formula for "norm" and we get the same result. 
> Suppose the units in the x are in mts, then np.linalg.norm(x), then our vector x is of length 93.12m.
